## 3.Graficas 

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.lmplot(x="Year", y="Value", data=WORLD_MINF,col="CountryName")
```

Realizando las regresiones lineales de los diferentes lugares entre el tiempo y el valor del indicador de mortalidad, nos arrojan  de manera visual una relación lineal entre ellos y esto nos indica que a través del tiempo y de la historia tanto política como económica de estos, han ido cambiando o manteniéndose un desarrollo estable o mejor. 

```python
flights_long = WORLD_MINF
flights = flights_long.pivot("CountryName", "Year", "Value")

f, ax = plt.subplots(figsize=(18, 18))
sns.heatmap(flights, annot=False, linewidths=.5, ax=ax)
```

Con este mapa de tiempo entre los lugares con el indicador de mortalidad de interés, nos damos cuenta de que según el color ha sido el avance en ellos, siendo el color más claro un desarrollo bajo mientras que el obscuro arroja uno mejor, y como lo mencionamos anterior mente, aquí podemos observar de una mejor manera que estos lugares han tendido ha contar con un mejor desarrollo según sea el caso.

```python
f, ax = plt.subplots(figsize=(14, 5))
g=sns.violinplot( x='CountryName',y='Value', data=WORLD_MINF,scale="width",ax=ax)
plt.xticks(rotation=45)
```

Con el grafico anterior podemos ver como el valor de su indice de mortalidad en cada region se ha distribuido conforme han pasado los años, vemos que en paises como Norte America siempre ha sido muy bajo su indice de mortalidad, lo que nos indica que no hahabido una diferencia muy notable, pero que se ha mantenido conforme los años pasan.

```python
#Se crea este dataframe para poder tener la siguiente tabla y poder hacer las gráficas

ARBM=ARB[ARB['IndicatorCode']=='SP.DYN.IMRT.IN'] 
ARBM=ARBM.drop(['CountryName','IndicatorName','IndicatorCode','CountryCode'],axis=1)
ARBM.rename(columns = {'Value':'ARB'}, inplace = True) 

MEAM=MEA[MEA['IndicatorCode']=='SP.DYN.IMRT.IN'] 
MEAM=MEAM.drop(['CountryName','IndicatorName','IndicatorCode','CountryCode'],axis=1)
MEAM.rename(columns = {'Value':'MEA'}, inplace = True) 

EASM=EAS[EAS['IndicatorCode']=='SP.DYN.IMRT.IN'] 
EASM=EASM.drop(['CountryName','IndicatorName','IndicatorCode','CountryCode'],axis=1)
EASM.rename(columns = {'Value':'EAS'}, inplace = True) 

ECSM=ECS[ECS['IndicatorCode']=='SP.DYN.IMRT.IN'] 
ECSM=ECSM.drop(['CountryName','IndicatorName','IndicatorCode','CountryCode'],axis=1)
ECSM.rename(columns = {'Value':'ECS'}, inplace = True) 


LCNM=LCN[LCN['IndicatorCode']=='SP.DYN.IMRT.IN'] 
LCNM=LCNM.drop(['CountryName','IndicatorName','IndicatorCode','CountryCode'],axis=1)
LCNM.rename(columns = {'Value':'LCN'}, inplace = True) 


NACM=NAC[NAC['IndicatorCode']=='SP.DYN.IMRT.IN'] 
NACM=NACM.drop(['CountryName','IndicatorName','IndicatorCode','CountryCode'],axis=1)
NACM.rename(columns = {'Value':'NAC'}, inplace = True) 


SASM=SAS[SAS['IndicatorCode']=='SP.DYN.IMRT.IN'] 
SASM=SASM.drop(['CountryName','IndicatorName','IndicatorCode','CountryCode'],axis=1)
SASM.rename(columns = {'Value':'SAS'}, inplace = True) 

SSFM=SSF[SSF['IndicatorCode']=='SP.DYN.IMRT.IN'] 
SSFM=SSFM.drop(['CountryName','IndicatorName','IndicatorCode','CountryCode'],axis=1)
SSFM.rename(columns = {'Value':'SSF'}, inplace = True) 

a = pd.merge(ARBM, MEAM, on='Year')
b = pd.merge(EASM, ECSM, on='Year')
c = pd.merge(LCNM, NACM, on='Year')
d = pd.merge(SASM, SSFM, on='Year')
A = pd.merge(a,b, on='Year')
B = pd.merge(c,d, on='Year')
result = pd.merge(A,B, on='Year')
result.to_csv('Tabla_Mortalidad_Filas_years_colum_pais.csv', index=False)
```

```python

x=result.drop(['Year'],axis=1)
x.plot.bar(stacked=True)
```

En el grafico anterior podemos observar como el indice de mortalidad infantil en el conjunto de las 8 regiones a anilizar se fue comportando y aminorando conforme fueron pasando los años, podemos notar claramente como es que todas las regiones han ido mejorando notablemente con el paso de los años, sin embargo de la misma manera podemos comparar entre las regiones la diferencia de muertes infantiles en promedio.
